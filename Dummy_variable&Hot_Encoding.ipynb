```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("homeprices (1).csv")
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merge=pd.concat([dummies,df],axis="columns")
merge
```

```python
final=merge.drop(["town","west windsor"],axis="columns")
final
```

```python
from sklearn import linear_model
x=final.drop("price",axis="columns")
y=final.price
x
```

```python
ref=linear_model.LinearRegression()
ref.fit(x,y)
```

```python
ref.predict([[0,0,2700]])
```

```python
d=pd.read_csv("carprices (1).csv")
d
```

```python
dummies=pd.get_dummies(d.Car_Model)
dummies
```

```python
merge=pd.concat([dummies,d],axis="columns")
merge
```

```python
final=merge.drop(["Mercedez Benz C class","Car_Model"],axis="columns")
final
```

```python
x=final.drop("Sell_Price",axis="columns")
x
```

```python
y=final.Sell_Price
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ref=LinearRegression()
ref.fit(x,y)
```

```python
ref.predict([[0,0,45000,4]])
```

```python
ref.predict([[0,1,86000,7]])
```

```python
ref.score(x,y)
```
