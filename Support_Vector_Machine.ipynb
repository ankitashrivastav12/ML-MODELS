```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris=load_iris()
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df["target"]=iris.target
```

```python
df["flower_name"]=df.target.apply(lambda x:iris.target_names[x])
df
```

```python
df[df.target==0]
```

```python
df[df.target==2]
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],color="green",marker="+")
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],color="blue",marker=".")
plt.scatter(df2["sepal length (cm)"],df2["sepal width (cm)"],color="red",marker="+")
```

```python
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm))")
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"],color="green",marker="+")
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color="blue",marker=".")
plt.scatter(df2["petal length (cm)"],df2["petal width (cm)"],color="red",marker="+")
```

```python
x=df.drop(["target","flower_name"],axis="columns")
y=df.flower_name
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=SVC()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[5.1,3.7,1.5,0.4]])
```

```python
model_C=SVC(C=1000)
model_C.fit(x_train,y_train)
model_C.score(x_test,y_test)
```

```python
model_g=SVC(gamma=10)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```

```python
model_linear_kernel=SVC(kernel="linear")
model_linear_kernel.fit(x_train,y_train)
model_linear_kernel.score(x_test,y_test)
```

```python
model_rbf_kernel=SVC(kernel="rbf")
model_rbf_kernel.fit(x_train,y_train)
model_rbf_kernel.score(x_test,y_test)
```

```python
model_sigmoid_kernel=SVC(kernel="sigmoid")
model_sigmoid_kernel.fit(x_train,y_train)
model_sigmoid_kernel.score(x_test,y_test)
```

```python
model_polynomial_kernel=SVC(kernel="poly")
model_polynomial_kernel.fit(x_train,y_train)
model_polynomial_kernel.score(x_test,y_test)
```
