```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("income.csv")
df
```

```python
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.scatter(df.Age,df["Income($)"],color="red")
```

```python
model=KMeans(n_clusters=3)
y_predict=model.fit_predict(df[["Age","Income($)"]])
y_predict
```

```python
df["cluster"]=y_predict
df
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.scatter(df0.Age,df0["Income($)"],color="red")
plt.scatter(df1.Age,df1["Income($)"],color="green")
plt.scatter(df2.Age,df2["Income($)"],color="blue")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[["Age"]])
df["Age"]=scaler.transform(df[["Age"]])
scaler.fit(df[["Income($)"]])
df["Income($)"]=scaler.transform(df[["Income($)"]])
df
```

```python
model=KMeans(n_clusters=3)
y_predict=model.fit_predict(df[["Age","Income($)"]])
df["cluster"]=y_predict
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.scatter(df0.Age,df0["Income($)"],color="red")
plt.scatter(df1.Age,df1["Income($)"],color="green")
plt.scatter(df2.Age,df2["Income($)"],color="blue")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color="black",marker="*",label="centeroid")
```

```python
sse=[]
k_rng= range(1,10)
for k in k_rng:
  model=KMeans(n_clusters=k)
  model.fit(df[["Age","Income($)"]])
  sse.append(model.inertia_)
sse
```

```python
plt.xlabel("K")
plt.ylabel("SUM OF SQUARED ERROR")
plt.plot(k_rng,sse)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
plt.xlabel("PETAL LENGTH")
plt.ylabel("PETAL WIDTH")
plt.scatter(df["petal length (cm)"],df["petal width (cm)"],color="green")
```

```python
km=KMeans(n_clusters=2)
y_predict=km.fit_predict(df[["petal length (cm)","petal width (cm)"]])
y_predict
```

```python
df["cluster"]=y_predict
df
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
```

```python
plt.xlabel("PETAL LENGTH")
plt.ylabel("PETAL WIDTH")
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"],color="green")
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color="red")
```

```python
scaler=MinMaxScaler()
scaler.fit(df[["petal length (cm)"]])
df["petal length (cm)"]=scaler.transform(df[["petal length (cm)"]])
scaler.fit(df[["petal width (cm)"]])
df["petal width (cm)"]=scaler.transform(df[["petal width (cm)"]])
df
```

```python
km=KMeans(n_clusters=2)
y_predict=km.fit_predict(df[["petal length (cm)","petal width (cm)"]])
df["cluster"]=y_predict
df0=df[df.cluster==0]
df1=df[df.cluster==1]
plt.xlabel("PETAL LENGTH")
plt.ylabel("PETAL WIDTH")
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"],color="green")
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color="red")
```

```python
sse=[]
k_rng= range(1,10)
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[["petal length (cm)","petal width (cm)"]])
  sse.append(km.inertia_)
sse
```

```python
plt.xlabel("K")
plt.ylabel("sum of squared error")
plt.plot(k_rng,sse)
```
