```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img('/content/bulldog.jpg',target_size=(200,200))
```

```python
type(img)
```

```python
img
```

```python
datagen=ImageDataGenerator(rotation_range=30, width_shift_range= 0.2, height_shift_range= 0.2, shear_range=0.2, zoom_range= 0.2, horizontal_flip=True,vertical_flip= False,fill_mode='constant')
```

```python
img=image.img_to_array(img)
type(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='/content/drive/MyDrive/aug'):
  i += 1
  if i == 10:
    break
```
